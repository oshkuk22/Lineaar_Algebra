__1.__ Решить систему уравнений методом Крамера:

   а) $\begin{cases}
x_{1}-2x_{2}=1 \\
3x_{1}-4x_{2}=7
\end{cases}$
    
   б) $\begin{cases}
2x_{1}-x_{2}+5x_{3}=10 \\
x_{1}+x_{2}-3x_{3}=-2 \\
2x_{1}+4x_{2}+x_{3}=1
\end{cases}$

__Решение:__


```python
#a)
import numpy as np

delta_ = np.linalg.det(np.array([[1, -2],[3, -4]]))

delta_1 = np.linalg.det(np.array([[1, -2],[7, -4]]))

delta_2 = np.linalg.det(np.array([[1, 1],[3, 7]]))

x_1 = round(delta_1 / delta_, 1)

x_2 =round(delta_2 / delta_, 1)

print(x_1, x_2)
```

__5*.__ Написать на Python программу с реализацией одного из изученных алгоритмов решения СЛАУ.

```python
import numpy as np

A = np.array([[2, -1, 5], [1, 1, -3], [2, 4, 1]])
b = np.array([10, -2, 1])

def kramer_(A, b):
    solving = []
    if A.shape[0] != A.shape[1]:
        return 'Матрица не квадратная'
    elif A.shape[0] != b.shape[0]:
        return 'Количество свободных членов не равно кличеству уравнений'
    else:
        delta_ = np.linalg.det(A)
        
        for i in range(A.shape[1]):
            if i == 0:
                delta_new = np.linalg.det(np.column_stack([b,A[:, 1:]]))
                solving.append(round(delta_new / delta_, 1))
            elif i == A.shape[1]-1:
                delta_new = np.linalg.det(np.column_stack([A[:,:A.shape[1]-1], b]))
                solving.append(round(delta_new / delta_, 1))
            else:
                delta_new = np.linalg.det(np.column_stack([A[:, :i], b, A[:, (i + 1):(A.shape[1])]]))
                solving.append(round(delta_new / delta_, 1))
    return solving                



print(kramer_(A,b))
```
