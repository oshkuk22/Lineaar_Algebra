__2.__ Дан оператор поворота на 180 градусов, задаваемый матрицей 

$$A=\begin{pmatrix}
-1 & 0\\ 
0 & -1
\end{pmatrix}.$$

Показать, что __любой__ вектор является для него собственным.

__Решение:__
    
$\begin{pmatrix}
-1 & 0\\ 
0 & -1
\end{pmatrix}\cdot \begin{pmatrix}
x1\\ 
x2
\end{pmatrix} = \lambda \cdot \begin{pmatrix}
x1\\ 
x2
\end{pmatrix}$

$\begin{cases}
-x_{1}= \lambda x_{1}, \\
 -x_{2}= \lambda x{2}.
\end{cases} \Rightarrow 
\begin{cases}
-x_1= -x_1, \\
 -x_2= -x_2
\end{cases}\Rightarrow$ любой вектор является собственным

__3.__ Пусть линейный оператор задан матрицей

$$A=\begin{pmatrix}
1 & 1\\ 
-1 & 3
\end{pmatrix}.$$

Установить, является ли вектор $x=(1,1)$ собственным вектором этого линейного оператора.

__Решение:__

$\begin{pmatrix}
1 & 1\\ 
-1 & 3
\end{pmatrix}\cdot \begin{pmatrix}
x1\\ 
x2
\end{pmatrix} = \lambda \cdot \begin{pmatrix}
x1\\ 
x2
\end{pmatrix}$

$\begin{cases}
x_{1}+x_{2}= \lambda x_{1}, \\
 -x_{1}+3x_{2}= \lambda x{2}.
\end{cases} \Rightarrow 
\begin{cases}
1+1= \lambda, \\
 -1+3= \lambda 1.
\end{cases}\Rightarrow 
\begin{cases}
\lambda = 2, \\
\lambda = 2.
\end{cases}\Rightarrow$ вектор является собственным
